```python
## Task 1: Mark all launch sites on a map
# =================================================================
# COMPLETE CODE FOR TASK 1: MAPPING LAUNCH SITES
# =================================================================

# --- Step 1: Install Libraries (using piplite for this environment) ---
import piplite
await piplite.install(['pandas'])
await piplite.install(['folium'])


# --- Step 2: Import Libraries ---
import pandas as pd
import folium
# Import special libraries for loading data in this environment
from pyodide.http import pyfetch
import io


# --- Step 3: Load the Dataset from a Direct URL ---
# This is the correct, direct link to the raw CSV file
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'

# Define an async function to fetch the data
async def download(url, filename):
    response = await pyfetch(url)
    if response.status == 200:
        with open(filename, "wb") as f:
            f.write(await response.bytes())

# Download the file and read it into a pandas DataFrame
await download(URL, "spacex_launch_geo.csv")
spacex_df = pd.read_csv("spacex_launch_geo.csv")


# --- Step 4: Prepare Data for Mapping ---
# Select relevant columns and find unique launch sites
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]


# --- Step 5: Create the Map and Add Markers ---
# Initialize the map, centered around a general launch area
nasa_coordinate = [29.55968488852867, -95.0830971930759] # A general center point
site_map = folium.Map(location=nasa_coordinate, zoom_start=4)

# Create a feature group for the launch sites
launch_sites_group = folium.map.FeatureGroup()

# Loop through each launch site and add a CircleMarker to the map
for index, site in launch_sites_df.iterrows():
    # Define coordinates and name for each site
    coordinate = [site['Lat'], site['Long']]
    site_name = site['Launch Site']
    
    # Create a circle marker
    circle = folium.Circle(
        coordinate,
        radius=1000, # Radius in meters
        color='#d35400',
        fill=True
    ).add_child(folium.Popup(site_name)) # Add a popup with the site name
    
    # Create a marker with an icon
    marker = folium.map.Marker(
        coordinate,
        # Create an icon as a text label
        icon=folium.DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12; color:#d35400;"><b>{site_name}</b></div>',
        )
    )
    site_map.add_child(circle)
    site_map.add_child(marker)

# Display the map
site_map
```

```python
# Task 2: Mark the success/failed launches for each site on the map
# Import the MarkerCluster library
from folium.plugins import MarkerCluster

# Create a new map
marker_cluster_map = folium.Map(location=nasa_coordinate, zoom_start=4)

# Create a new MarkerCluster object
marker_cluster = MarkerCluster()

# Add the MarkerCluster to the map
marker_cluster_map.add_child(marker_cluster)

# Function to assign color to launch outcome
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green' # Green for success
    else:
        return 'red'   # Red for failure

# Apply the function to create a new 'marker_color' column
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)

# Iterate over each row in the spacex_df dataframe
for index, record in spacex_df.iterrows():
    # Create a Marker object for each launch record
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        icon=folium.Icon(color='white', icon_color=record['marker_color']),
        popup=record['Launch Site']
    )
    # Add the marker to the MarkerCluster
    marker_cluster.add_child(marker)

# Display the map
marker_cluster_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import sin, cos, sqrt, atan2, radians

# --- Step 1: Define the distance calculation function (Haversine formula) ---
def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# --- Step 2: Define coordinates for a launch site and a proximity point ---
# Let's find the coordinates of KSC LC-39A
ksc_lat = 28.573255
ksc_lon = -80.646895

# And a point on the nearby coastline
coastline_lat = 28.56367
coastline_lon = -80.57163

# --- Step 3: Calculate the distance and create the map elements ---
# Calculate the distance
distance_to_coastline = calculate_distance(ksc_lat, ksc_lon, coastline_lat, coastline_lon)

# Create a Folium map centered on the launch site
distance_map = folium.Map(location=[ksc_lat, ksc_lon], zoom_start=14)

# Add a marker for the coastline point
folium.Marker(
    [coastline_lat, coastline_lon],
    icon=folium.DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_to_coastline),
    )
).add_to(distance_map)

# Create a PolyLine to draw a line between the launch site and the coastline
points = [[ksc_lat, ksc_lon], [coastline_lat, coastline_lon]]
folium.PolyLine(points, color='blue').add_to(distance_map)

# Display the map
distance_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site

# Launch Site Coordinates (e.g., CCAFS SLC-40)
launch_site_lat = 28.563197
launch_site_lon = -80.576820
launch_site_coords = [launch_site_lat, launch_site_lon]

# Proximity Points Coordinates
proximities = {
    'Coastline': [28.56367, -80.57163],
    'Railway': [28.57205, -80.58525],
    'Highway': [28.56338, -80.57079],
    'City (Cape Canaveral)': [28.40161, -80.6042]
}

# Create a new map centered on the launch site
proximity_map = folium.Map(location=launch_site_coords, zoom_start=12)

# Add a marker for the launch site itself for clarity
folium.Marker(launch_site_coords, popup='Launch Site CCAFS SLC-40').add_to(proximity_map)

# Loop through each proximity point
for name, coords in proximities.items():
    # 1. Calculate the distance
    distance = calculate_distance(launch_site_lat, launch_site_lon, coords[0], coords[1])
    
    # 2. Create a distance marker
    distance_marker = folium.Marker(
       coords,
       icon=folium.DivIcon(
           icon_size=(100,20), # Increased size for better visibility
           icon_anchor=(0,0),
           html='<div style="font-size: 12; color:#d35400;"><b>{} {:10.2f} KM</b></div>'.format(name, distance),
           )
       )
    proximity_map.add_child(distance_marker)
    
    # 3. Draw a PolyLine
    line_points = [launch_site_coords, coords]
    line = folium.PolyLine(locations=line_points, weight=1, color='red')
    proximity_map.add_child(line)

# Display the final map with all lines and distances
proximity_map
```
